
### An introduction to images

The code below can be used on any image (png, jpg) to save a text-based file that has three columns: Red, Green and Blue, with each row
representing the colour values of a pixel in the image. You should load the image into the local workspace, and change the image input and file
output names to suit yourself. You can then download the text file for use elsewhere. You may find the information further down this notebook
helpful to understand image files.

```python
import numpy as np
import cv2 as cv
import csv
# change file names here
image_name_in = "/content/WTflyeye(1)png (1).png"
file_name_out = "/content/WTflyeye cv (1).csv"
# imread assumes BGR colour space
image = cv.imread(image_name_in)
RGB_img = cv.cvtColor(image, cv.COLOR_BGR2RGB) # convert to RGB
width, height, colourchannels = image.shape # size of array
import matplotlib.pyplot as plt
print(image[30][45]) # BGR order
print(RGB_img[30][45]) # RGB order
plt.imshow(RGB_img) # plot the image
plt.title('RGB Image'), plt.xticks([]), plt.yticks([]) # give the plot a title, and suppress axis ticks
plt.show() # show the plot
# save the channels out
with open(file_name_out, 'w', newline='') as csvfile:
  filewriter = csv.writer(csvfile) # create the csv writer - filewriter is just an informative name, you could use anything
  filewriter.writerow(['Red', 'Green', 'Blue'])
  for x in range(width):
    for y in range(height):
      filewriter.writerow(RGB_img[x,y])
```

### Produces an annotated image of the red, green and blue channels (to the screen and/or to file);

```python
def plotHistogram(a):
    """
    Plot histogram of RGB Pixel Intensities
    """
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.imshow(a)
    plt.axis('off')
    histo = plt.subplot(1,2,2)
    histo.set_ylabel('Count')
    histo.set_xlabel('Pixel Intensity')
    n_bins = 30
    plt.hist(a[:,:,0].flatten(), bins= n_bins, lw = 0, color='r', alpha=0.5);
    plt.hist(a[:,:,1].flatten(), bins= n_bins, lw = 0, color='g', alpha=0.5);
    plt.hist(a[:,:,2].flatten(), bins= n_bins, lw = 0, color='b', alpha=0.5);
plotHistogram(image)
```
